```python
import os
```

```python
def iterate_dirs(dir, func):
    for f in os.scandir(dir):
        if f.is_file():
            func(f)
        elif f.name[-6:] != '.files':
            iterate_dirs(f, func)
```

```python
def print_name(f):
    print(f.name)

def filter_name(f):
    return f.name[-4:] == '.htm'

def read(f, encoding='gbk'):
    file = open(f, encoding=encoding)
    lines = file.readlines()
    file.close()
    return lines

def write(f, lines):
    file = open(f, 'w', encoding='utf8')
    file.writelines(lines)
```

```python
f = "files/___left.htm"
print(read(f))
print(read(f, 'utf8'))
```

```python
def do_job(f):
    if filter_name(f):
        print_name(f)
        lines = read(f)
        write(f, lines)
```

```python
iterate_dirs("./files", do_job)
```
